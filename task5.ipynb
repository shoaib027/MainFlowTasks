```python
#1 People having heart disease or not
data.target.value_counts()
```

```python
data.target.value_counts().plot(kind = "bar", color = ["red", "green"])
plt.title("Number of people having heart disease or not")
plt.xlabel("1= Heart Disease, 0 Not having Heart Disease")
plt.ylabel("Number of people")
plt.xticks(rotation=0);
```

```python
data.sex.value_counts()
```

```python
data.sex.value_counts().plot(kind = "pie", figsize=(8,6))
plt.title('Male Female ratio')
plt.legend(['Male', 'Female']);
```

```python
#2 People of which sex has most heart disease
pd.crosstab(data.target, data.sex)
```

```python
sns.countplot(x='target', data = data, hue='sex')
plt.title('Heart Disease Frequency for sex')
plt.xlabel('0 = No Heart Disease, 1 = Heart Disease');
```

```python
#3 People of which sex has which type of chest pain
data.cp.value_counts()
```

```python
data.cp.value_counts().plot(kind='bar', color = ['hotpink', 'yellow', 'green', 'orange'])
plt.title('Chest pain type vs count')
```

```python
pd.crosstab(data.sex,data.cp).plot(kind ='bar', color = ['hotpink', 'yellow', 'green', 'orange'])
plt.title('Type of chest pain for sex')
plt.xlabel('0 =  Female, 1 = Male');
```

```python
#4 People with which chest pain are more prone to heart disease
pd.crosstab(data.cp, data.target)
sns.countplot(x = 'cp', data = data, hue = 'target');
```

```python
sns.displot(x = 'age', data = data, bins = 30, kde = True);
```
