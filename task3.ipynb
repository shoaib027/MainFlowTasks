```python
import pandas as pd
import matplotlib. pyplot as plt
```

```python
data = pd.read_csv("files/householdtask3.csv")
```

```python
data
```

```python
plt.bar(data['year'], data['tot_hhs'])
plt.title('Total number of households over the years')
plt.xlabel('Years')
plt.ylabel('Total Households')
plt.show()
```

```python
# Aggregate the values
aggregated_df = data.groupby('income').median().reset_index()

# Plot
plt.figure(figsize = (12, 4)) 
plt.plot(aggregated_df['income'], aggregated_df['expenditure'], marker='o', color='orange')
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.legend(['Expenditure'])
plt.title('Median expenditure')
plt.show()
```
